```python
import tensorflow as tf
import numpy as np
```

```python
tf.__version__
```

输入输出

```python
x1 = np.array([1,2,3,4,5],dtype=float)
y1 = np.array([101,201,301,401,501],dtype=float)
```

定义网络结构

```python
l0 = tf.keras.layers.Dense(input_shape=[1],units=1)
```

```python
#model = tf.keras.Sequential([l0])


model = tf.keras.Sequential([
    tf.keras.layers.Dense(input_shape=[1],units=4),
    tf.keras.layers.Dense(input_shape=[4],units=4),
    tf.keras.layers.Dense(input_shape=[4],units=1)
])
```

编译(loss optimizer)

```python
model.compile(loss='mean_squared_error',optimizer=tf.keras.optimizers.Adam(learning_rate=0.01))
```

训练(fit)

```python
history = model.fit(x1,y1,epochs=1500,verbose=True)
```

```python
model.predict([7])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['loss'])
```
